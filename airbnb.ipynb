```python
import pandas as pd
```

```python
df = pd.read_csv("airbnb_sales_data.csv")
df.head()
```

```python
df.info()
```

```python
df['Construction year'] = pd.to_numeric(df['Construction year'], errors = 'coerce')
```

```python
df.info()
```

```python
df = df.drop_duplicates().reset_index(drop=True)
df['listing_id'] = df.index
```

```python
df.head()
```

```python
dim_host = df[['host id', 'host name', 'host_identity_verified']].reset_index(drop=True)
```

```python
dim_host['dim_host_id'] = dim_host.index
```

```python
dim_host
```

```python
dim_room_type = df[['room type']].reset_index(drop=True)
dim_room_type['room_type_id'] = dim_room_type.index
dim_room_type
```

```python
dim_cancellation_policy = df[['cancellation_policy']].reset_index(drop=True)
dim_cancellation_policy['cancellation_policy_id'] = dim_cancellation_policy.index
dim_cancellation_policy.head()
```

```python
spelling_corrections = {
    'brookln': 'Brooklyn',
    'brooklyn': 'Brooklyn',
    'manhattan' : 'Manhattan',
    'queens' : 'Queens',
    'bronx' : 'Bronx',
    'statens island' : 'Staten Island'
}
df['neighbourhood group'] = df['neighbourhood group'].replace(spelling_corrections)
df.head()
```

```python
borough_neighbourhood_id = {
    'Brooklyn' : 1,
    'Manhattan' : 2,
    'Queens' : 3,
    'Bronx' : 4,
    'Staten Island' : 5
}

df['neighbourhood_group_id'] = df['neighbourhood group'].map(borough_neighbourhood_id)
df.head()
    
dim_neighbourhood = df[['neighbourhood group', 'neighbourhood_group_id', 'neighbourhood', 'lat', 'long', 'country', 'country code']].reset_index(drop=True)
dim_neighbourhood['neighbourhood_id'] = dim_neighbourhood.index
dim_neighbourhood.head()
```

```python
dim_listing_details = df[['listing_id', 'NAME', 'house_rules', 'license']].reset_index(drop=True)
dim_listing_details.head()
```

```python
dim_listing_details= dim_listing_details.dropna(axis=1, how = 'all')
dim_listing_details.head()
```

```python
# 1️⃣ Create safe lookups for each dimension (drop duplicates on the business key)
room_type_lookup = dim_room_type[['room type', 'room_type_id']].drop_duplicates(subset=['room type'])
cancellation_lookup = dim_cancellation_policy[['cancellation_policy', 'cancellation_policy_id']].drop_duplicates(subset=['cancellation_policy'])
neighbourhood_lookup = dim_neighbourhood[['neighbourhood group', 'neighbourhood', 'neighbourhood_id']].drop_duplicates(subset=['neighbourhood group', 'neighbourhood'])

# 2️⃣ Build the fact table in one go
fact_listings = (
    df.copy()
    # merge room type
    .merge(room_type_lookup, on='room type', how='left')
    # merge cancellation policy
    .merge(cancellation_lookup, on='cancellation_policy', how='left')
    # merge neighbourhood
    .merge(neighbourhood_lookup, on=['neighbourhood group', 'neighbourhood'], how='left')
)[[
    'listing_id',                     # from df
    'host id',                        # natural key
    'neighbourhood_id',
    'room_type_id',
    'cancellation_policy_id',
    'price',
    'service fee',
    'minimum nights',
    'number of reviews',
    'reviews per month',
    'review rate number',
    'calculated host listings count',
    'availability 365',
    'Construction year',
    'last review',
    'instant_bookable'
]].drop_duplicates()


```
